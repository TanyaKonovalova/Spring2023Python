ЗАДАНИЕ №1

Рассморим задачу на полиномиальную регрессию!

```python
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def approx(n):
    dots = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)
    w = np.linalg.solve(a, b)  #точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #решение полиномиальной регрессии
    return res


x = np.linspace(1, 15, 250)
plt.grid()
plt.title('График исходной функции:')
plt.plot(x, f(x), color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График точек, по которым строился многочлен при n = 5:')
plt.plot(np.linspace(1, 15, 5), f(np.linspace(1, 15, 5)), '8', color='green')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График полиномиальной аппроксимации при n = 5:')
plt.plot(np.linspace(1, 15, 5), approx(5), color='yellow')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

ЗАДАНИЕ №2

Рассморим задачу на полиномиальную регрессию с зашумлёнными данными!

```python
import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots) + np.random.normal(0, 0.01)
    w = np.linalg.solve(a, b)  #точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #решение полиномиальной регрессии с шумом
    return res

x = np.linspace(1, 15, 250)
plt.grid()
plt.title('График исходной функции:')
plt.plot(x, f(x), color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График точек, по которым строился многочлен при n = 5:')
plt.plot(np.linspace(1, 15, 5), f(np.linspace(1, 15, 5)), '8', color='green')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График полиномиальной аппроксимации с шумом при n = 5:')
plt.plot(np.linspace(1, 15, 5), approx(5), color='yellow')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

ЗАДАНИЕ №3

Рассморим задачу на метод наименьших квадратов!

```python
N = 100
n = 2
dots = np.linspace(1, 15, N)
a = np.array([[x**j for j in range(n)] for x in dots])
b = f(dots) + np.random.normal(0, 0.1)

w, resid, rank, s = numpy.linalg.lstsq(a, b)  #точки в полиноме
res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #решение МНК

x = np.linspace(1, 15, 250)
plt.grid()
plt.title('График исходной функции:')
plt.plot(x, f(x), color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График точек, по которым строился многочлен n = {n}:')
plt.plot(dots, b, '8', color='green')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()

plt.grid()
plt.title(f'График метода наименьших квадратов при n = {n}:')
plt.plot(dots, res, color='yellow')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

```python
print ('Ответ: Из графиков видно, что при малом n, например, n = 2, график МНК неточен. \
Однако при увеличении n, аппроксимация будет становится точнее, и график МНК будет приближаться к исходному!')
```
