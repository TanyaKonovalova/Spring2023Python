## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
#Пункт 1
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x1 = np.linspace(-1, 1, 250)
y1 = x1**2 * np.sin(1/(x1**2)) + x1

plt.title(1)
plt.grid()
plt.plot(x1, y1, color='black')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')

plt.show()
```

```python
#Пункт 2
plt.title(2)
x2 = np.linspace(5, 25, 250)
y2 = abs(1 - (x2 ** 2)/(1 + x2 ** 2))
plt.grid()
plt.semilogy(x2, y2, color='red')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')

plt.show()
```

```python
#Пункт 3
from scipy.special import ellipk, wofz

plt.figure(figsize=(10, 5))
plt.title(3)
x3 = np.linspace(-5, 5, 100)
y3 = ellipk(x3) #Полный эллиптический интеграл первого рода
x4 = x3
y4 = wofz(x4) #Функция Фаддеева


plt.grid()
plt.plot(x3, y3, color='red', linestyle='--', label='Elliptic integral')
plt.plot(x4, y4, color='black', linestyle=':', label='The Faddeev Function')
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.legend()
plt.show()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6*np.pi, 6*np.pi, 250)
y = np.sin(x) / x

plt.figure(figsize=(16, 8))
plt.grid()
plt.plot(x, y)
labels = [(str(-12+i) + 'π/2' if i != 12 else '0') for i in range(25)]
plt.xticks(np.arange(-6*np.pi, 6*np.pi+np.pi/2, np.pi/2), labels)
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
#Пункт 1
x = np.arange(-np.pi, np.pi, 0.1)
y = x
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

plt.imshow(Z, cmap='viridis', extent=[-np.pi, np.pi, -np.pi, np.pi])
plt.colorbar()
plt.title(1)
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```

```python
#Пункт 2
x = np.linspace(-3, 3, 100)
y = x
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

plt.imshow(Z, cmap='Spectral', extent=[-3, 3, -3, 3])
plt.colorbar()
plt.title(2)
plt.xlabel('Ось: OX')
plt.ylabel('Ось: OY')
plt.show()
```
